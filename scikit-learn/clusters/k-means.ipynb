```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

%matplotlib inline
```

```python
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (12,8)
```

```python
X, y = make_blobs(n_samples=150, n_features=2, centers=4, random_state=1)
```

```python
plt.scatter(X[:,0], X[:,1])
```

```python
###Using of K-means method
kmeans = KMeans(n_clusters=4, init='k-means++', random_state=1, n_init=10)
```

```python
kmeans.fit(X)
```

```python
labels = kmeans.labels_
```

```python
plt.scatter(X[:,0], X[:,1], c=labels)
```

```python
###Chosing the best number of cluster by using of Elbow method

crit = []

for k in range(2,8):
    kmeans = KMeans(n_clusters=k, random_state=1)
    kmeans.fit(X)
    
    crit.append(kmeans.inertia_)
```

```python
###The best number of clusters
plt.plot(range(2,8), crit)
```
