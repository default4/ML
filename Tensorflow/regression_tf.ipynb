```python
import tensorflow as tf
```

```python
###House Price prediction for Boston Housing Dataset
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.boston_housing.load_data()
```

```python
###Data normalization
###Mean
mean = X_train.mean(axis=0)
###Standard deviation
std = X_train.std(axis=0)
X_train -= mean
X_train /= std
X_test -= mean
X_test /= std
```

```python
###Model creation
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(128, activation='relu', input_shape=(X_train.shape[1],)))
model.add(tf.keras.layers.Dense(1))
model.compile(optimizer='adam', loss='mse', metrics=['mae'])
```

```python
###Model training
model.fit(X_train, y_train, epochs=100, batch_size=1, verbose=2)
```

```python
###Network accuracy
mse, mae = model.evaluate(X_test, y_test, verbose=0)

print("The average absolute error (thousand dollars):", mae)
```

```python
###Using the network to predict property prices
pred = model.predict(X_test)

print("Predicted price:", pred[1][0], ", correct price:", y_test[1])
```
